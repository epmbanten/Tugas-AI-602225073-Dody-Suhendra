```python
import numpy as np
import matplotlib.pyplot as plt

def fuzzy_dingin(t):
    if t <= 15:
        return 1.0
    elif 15 < t < 25:
        return (25 - t) / (25 - 15)
    else:
        return 0.0

def fuzzy_sedang(t):
    if t <= 15 or t >= 35:
        return 0.0
    elif 15 < t <= 25:
        return (t - 15) / (25 - 15)
    elif 25 < t < 35:
        return (35 - t) / (35 - 25)
    else:
        return 0.0

def fuzzy_panas(t):
    if t <= 25:
        return 0.0
    elif 25 < t < 35:
        return (t - 25) / (35 - 25)
    else:
        return 1.0

# Menentukan Temperatur Input
temp_input = 32.5

# Menghitung Derajat Keanggotaan
mu_dingin = fuzzy_dingin(temp_input)
mu_sedang = fuzzy_sedang(temp_input)
mu_panas = fuzzy_panas(temp_input)

# Menampilkan (Print) Hasil
print(f"Hasil Fuzzifikasi untuk Temperatur {temp_input}°C:")
print(f"Dingin: {mu_dingin * 100:.0f}%")
print(f"Sedang: {mu_sedang * 100:.0f}%")
print(f"Panas : {mu_panas * 100:.0f}%")

# --- Visualisasi Grafik ---
t_range = np.linspace(10, 40, 500)
y_dingin = [fuzzy_dingin(x) for x in t_range]
y_sedang = [fuzzy_sedang(x) for x in t_range]
y_panas  = [fuzzy_panas(x) for x in t_range]

plt.figure(figsize=(10, 6))
plt.plot(t_range, y_dingin, 'b', label='Dingin')
plt.plot(t_range, y_sedang, 'g', label='Sedang')
plt.plot(t_range, y_panas, 'r', label='Panas')
plt.axvline(x=temp_input, color='brown', linestyle='--', label=f'Input {temp_input}°C')

plt.title('Gambar Tugas 2 - Fungsi Keanggotaan Temperatur')
plt.xlabel('Temperatur (t)')
plt.ylabel('Degree of Membership ($\mu$)')
plt.legend()
plt.grid(True)
plt.show()
```
